Sorting algorithms

```python
#sorting algorithms
import sys
sys.setrecursionlimit(3000)
import time
import random
#bubblesort

def bubblesort(keywords_list):
  for i in range(0, len(keywords_list)):
    for j in range(0,len(keywords_list) - 1):
      if keywords_list[j] [0] > keywords_list[j+1][0]:
        temp = keywords_list[j] + keywords_list [j+1]
        keywords_list[j+1] = temp

bubblesort(keywords_list)
print(keywords_list)

#insertion sort

def insertionsort(keywords_list):
  for i in range(0,len(keywords_list)):
    j = i
    while(j > 0 and keywords_list[j-1][0] > keywords_list[j][0]):
      temp = keywords_list[j]
      keywords_list[j] = keywords_list[j-1]
      keywords_list[j-1] = temp
      j = j-1

insertionsort(keywords_list)
print(keywords_list)

#quick sort

def partition(keywords_list, low, high):
  pivot = keywords_list[high]
  index = low

  for i in range(low, high):
    if keywords_list[i] <= pivot:
      temp = keywords_list[index]
      keywords_list[index] = keywords_list[i]
      keywords_list[i] = temp
      index = index + 1

  temp = keywords_list[high]
  keywords_list[high] = keywords_list[index]
  keywords_list[index] = temp

  return index

def qsort(keywords_list, low, high):
  if low < high:
    pivot = partition(keywords_list, low, high)

    qsort(keywords_list, low, pivot - 1);
    qsort(keywords_list, pivot + 1, high);

def quicksort(keywords_list):
    qsort(keywords_list, 0, len(keywords_list) - 1)

quicksort(keywords_list)
print(keywords_list)

#merge sort

def mergeSort(keywords_list):
    if len(keywords_list) > 1:
        mid = len(keywords_list) // 2
        left = keywords_list[:mid]
        right = keywords_list[mid:]

        mergeSort(left)
        mergeSort(right)

        i = 0
        j = 0
        k = 0

        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
              keywords_list[k] = left[i]
              i += 1
            else:
                keywords_list[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            keywords_list[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            keywords_list[k]=right[j]
            j += 1
            k += 1

mergeSort(keywords_list)
print(keywords_list)

start = time.time ()
albums = bubblesort(keywords_list)
end = time.time()
print(f"Bubble Sort: {(end - start)} seconds")
start = time.time()
albums = insertionsort(keywords_list)
end = time.time()
print(f"Insertion sort takes: {(end - start)} seconds")
start = time.time()
albums = quicksort(keywords_list)
end = time.time()
print(f"Quick sort takes: {(end - start)} seconds")
print(keywords_list)
start = time.time()
albums = mergeSort(keywords_list)
end = time.time()
print(f"Merge sort takes: {(end - start)} seconds")

print(keywords_list)
```
